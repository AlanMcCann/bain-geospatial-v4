```python
from time import sleep
from selenium import webdriver
# this is tested on Firefox or you can use "webdriver.Chrome()"
browser = webdriver.Chrome()
browser.get("https://mtg.bain.com")
sleep(5)
```

```python
for market_name in market_names:
    data_url = f'https://mtg.bain.com/globalassets/mtg/geospatialtool/v1/{market_name}_percent_data_v3.json'
    browser.get(data_url)
    sleep(1)
    if browser.page_source == NO_PAGE_MESSAGE:
        print(market_name, "failed", data_url)
```

```python
data_types = ["count",
"nshft",
"pshft",
"growth",
"percent",
"bg_settings"]
version = 3
for market_name in market_names:
    for data_type in data_types:
        data_url = f'https://mtg.bain.com/globalassets/mtg/geospatialtool/v1/{market_name}_{data_type}_data_v3.json'
        browser.get(data_url)
        sleep(1)
        if browser.page_source == NO_PAGE_MESSAGE:
            print(market_name, "failed", data_url)
```
